1. Create a function to print first 10 natural numbers

```python
print("-----First 10 natural numbers-----")
def NatNo(a=10):
    for i in range(1,a+1):
        print(i)
NatNo()
```

2. Create a function to calculate sum of first N natural numbers

```python
print("-----Sum of first N natural numbers-----")
def SumOfNatural(n):
    a = 0
    for i in range(1, n+1):
        a = a + i
    print(a)
SumOfNatural(6)
```

3. Create a function to reverse a number

```python
print("-----Reverse a number-----")
def RevNum(num):
    if num>0:
        ones_digit = num % 10
        tens_digit = (num - ones_digit)/10
        rev = ((ones_digit)*10) + tens_digit
        print(rev)
    else:
        print("Invalid input")
RevNum(-2)
RevNum(63)
```

4. Create a function to count digits in a number

```python
print("-----Count digits in a number-----")
def count_digits(num):
    num_str = str(num)
    num_len = len(num_str)
    print(f"Digits in the number: {num_len}")
count_digits(354964)
```

5. Create a function to check palindrome number

```python
print("-----Check palindrome number-----")
def ifpalindrome_num(num):
    num_str = str(num)
    rev_num = num_str[::-1]
    if (num_str == rev_num):
        print("Yes")
    else:
        print("No")
ifpalindrome_num(56965)
```

6. Create a function to generate Fibonacci series.

```python
print("-----Generate fibonacci series-----")
def fibonacci_series(num):
    a=0
    b=1
    c = a + b
    lis = [a, b, c]
    for i in range(num-3):
        a = b
        b = c
        c = a + b
        lis.append(c)
    print(lis)
fibonacci_series(6)
```

7. Calculator Using Functions that contains the following features;
    - User selects operation 
    - Program performs calculation 
    - Display result

```python
print("-----Calculator-----")
inp = str(input("What operation do you want to do, Addition(A), Subtraction(S), Multiplication(M) or Division(D)? "))
def calculator(num1,num2):
    ans = 0
    if inp == "A":
        ans = num1 + num2
        print(ans)
    elif inp == "S":
        ans = num1 - num2
        print(ans)
    elif inp == "M":
        ans = num1 * num2
        print(ans)
    elif inp == "D":
        ans = num1 / num2
        print(ans)
    else:
        print("Invalid input")
a = int(input("Enter the 1st number: "))
b = int(input("Enter the 2nd number: "))
print(f"The 1st number: {a}")
print(f"The 2nd number: {b}")
print(f"The operation: {inp}")
print(f"Result:", end=" ")
calculator(a,b)
```
