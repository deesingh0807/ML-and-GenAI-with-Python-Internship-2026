```python
#Convert temperature from Celsius to Fahrenheit.
TC = float(input("Enter the temperature in celsius: "))
TF = ((9/5)*TC) + 32
print("Temperature in Fahrenheit will be", TF)
```

```python
#Calculate average of 3 numbers.
no1 = float(input("Enter the 1st number: "))
no2 = float(input("Enter the 2nd number: "))
no3 = float(input("Enter the 3rd number: "))
Avg = (no1 * no2 * no3) / 3
print("Average of 3 numbers =", Avg)
```

```python
#Swap two numbers without third variable
a = float(input("Enter the 1st number: "))
b = float(input("Enter the 2nd number: "))
print(f"Before swapping: {a}, {b}")
a, b = b, a
print(f"After swapping: {a}, {b}")
```
