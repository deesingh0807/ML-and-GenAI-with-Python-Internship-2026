1. Find sum of first 10 natural numbers

```python
#Find sum of first 10 natural numbers
print("Sum of first 10 natural numbers")
a=0
for i in range(1,11):
    print(f"a = {a}, i = {i}")
    a = a + i
    print(f"a = a + i = {a}")
print(f"Sum of first 10 numbers: {a}")
```

2. Find factorial of a number

```python
#Find factorial of a number
print("Factorial of a number")
f = int(input("Enter the number whose factorial is to be found: "))
a = 1
for i in range(1,f+1):
    print(f"a = {a}, i = {i}")
    a = a * i
    print(f"a = a * i = {a}")
print(f"Factorial of the input number: {a}")
```

3. Print fibonacci Series

```python
#Print fibonacci series
j = int(input("Enter the number: "))
print(f"The number whose fibonacci series is to be printed out: {j}")
a = 0
b = 1
c = a + b
l =[a, b, c]
for i in range(2,j):
    a = b
    b = c
    c = a + b
    l.append(c)
print(l)
```
